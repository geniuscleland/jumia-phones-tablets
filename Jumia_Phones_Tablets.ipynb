```python
from urllib.request import Request,urlopen
# import request and urlopen libraries
```

```python
from bs4 import BeautifulSoup as bs
# import Beautiful Soup library
```

```python
url = 'https://www.jumia.com.gh/phones-tablets/'
req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})

# send request to url
```

```python
webpage = urlopen(req).read()
webpage
# open url response
```

```python
page_soup = bs(webpage, "html.parser")
```

```python
products = [pro.get_text() for pro in page_soup.select(".core .name")]
products
# get all product names of the page using the div class= core and h3 class = name
```

```python
price = [cost.get_text() for cost in page_soup.select(".core .prc")]
price
#  get all prices of products of the page using the div class= core and div class = prc
```

```python
links = [d.get('href') for d in page_soup.select(".core")]
links
```

```python
import pandas as pd
# import pandas library
PhoneTablets = pd.DataFrame({
    "Product Name": products[:130],
    "Price": price[:130],
    "Product Link":links[:130]
    
})
PhoneTablets
```

```python
PhoneTablets.to_csv("Jumia Phones and Tablets Assigment From Godwin Cleland.csv")
```
